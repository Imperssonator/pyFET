```python
import pandas as pd
import numpy as np
import os

rootdir = '/Users/Imperssonator/Dropbox/OFET Fab Table/OFET-Database-master/Papers/'

df = pd.DataFrame()

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        if file[-4:] == 'xlsx':
            dfi = pd.read_excel(os.path.join(subdir,file),index_col=0,header=None)
            dfi = dfi.transpose()
            df = df.append(dfi)
            print(file)
```
